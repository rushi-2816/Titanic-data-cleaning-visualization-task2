```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

%matplotlib inline

```

```python
df = pd.read_csv('train_sample.csv')  # Make sure this matches the uploaded file name
df.head()
```

```python
print("Shape of dataset:", df.shape)
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
plt.figure(figsize=(8, 5))
df['Age'].hist(bins=10, edgecolor='black', color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(x=df['Fare'], color='lightgreen')
plt.title('Fare Boxplot')
plt.show()
```

```python
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival by Gender')
plt.show()
```

```python
# Filter to only numeric columns
numeric_df = df.select_dtypes(include='number')

# Plot the correlation heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
fig = px.scatter(df, x='Age', y='Fare', color='Survived', title='Age vs Fare by Survival')
fig.show()
```
